```python
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
data = np.random.rayleigh(scale = 1.0, size = 1000)
```

```python
plt.figure(figsize= (6,6))
plt.hist(data, bins=100, edgecolor='black', alpha = 0.7)


plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 1k Rayleigh distributed Random Numbers")


plt.savefig("rayleigh_histogram.pdf")

plt.show()


```
