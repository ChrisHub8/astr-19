```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

```

```python
# Creating an array x ranging from 0 to 1 with 100 steps
x = np.linspace(0, 1, 100)
```

```python
def exp_function(x):
    return np.exp(x)
```

```python
# using function to compute y 
y = exp_function(x)


# plot x vs y

plt.plot(x,y)
plt.xlabel("Time[milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Session 7")
plt.grid()



plt.savefig("awesome_plot.pdf")
plt.show()
```
