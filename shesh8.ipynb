```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

```python
# creating a np array ranging from 0 to 1 with 100 steps
x = np.linspace(0,1,100)
```

```python
# define a function that returns both sinx and cosx
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)

```

```python
# compute y values

y_sin = sin_function(x)
y_cos = sin_function(x)


# creating a figure with 1 row and 2 columns 

fig, axes  = plt.subplots(1,2)

# left panel : sin(x) vs x 

axes[0].plot(x,y_sin)
axes[0].set_title("sin(x)")
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].grid(True)


# Right panel: cos(x) vs x 

axes[1].plot(x,y_cos)
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].grid(True)

# adjusting layout
plt.tight_layout()

# Save the figure as a pdf 
plt.savefig('sin_cos_plot.pdf')
plt.show()
```
